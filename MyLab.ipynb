```python
import pandas as pd
import numpy as np

x = np.arange(-10, 11)
df = pd.DataFrame({"x": x, "y": x ** 2})
df.to_csv("data.csv", index=False)
print(df)
```

```python
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")
plt.plot(df["x"], df["y"])
plt.xlabel("x")
plt.ylabel("f(x) = x²")
plt.title("y = x²")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Softmax function
def softmax(x):
    e_x = np.exp(x - np.max(x))  # subtract max for numerical stability
    return e_x / e_x.sum()

# Generate input values
x = np.linspace(-4, 4, 20)
y = softmax(x)

# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart
axes[0].bar(range(len(x)), y, color='steelblue', edgecolor='black', alpha=0.8)
axes[0].set_title('Softmax Output (Bar Chart)', fontsize=14)
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Probability')

# Line chart
axes[1].plot(x, y, color='tomato', linewidth=2.5, marker='o', markersize=5)
axes[1].set_title('Softmax Output vs Input (Line Chart)', fontsize=14)
axes[1].set_xlabel('Input value')
axes[1].set_ylabel('Probability')
axes[1].grid(True, linestyle='--', alpha=0.6)

plt.suptitle('Softmax Function', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
